```python
#Filling the null values

df2= df.fillna(value=0)
df2
```

```python
#distribution of legit transaction & fraudulent transaction
df2['Class'].value_counts()
```

```python
# seperating the data for analysis
legit = df2[df2.Class == 0]
fraud = df2[df2.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
#Amount column is the money involved in the transaction
```

```python
fraud.Amount.describe()
#mean is more in fraud than legit
```

```python
# Convert relevant columns to numeric if they contain strings
for col in df2.columns:
    if df2[col].dtype == 'object':
        try:
            df2[col] = pd.to_numeric(df2[col], errors='coerce')
        except ValueError:
            print(f"Could not convert column {col} to numeric")

# Calculate the mean after conversion
df2.groupby('Class').mean()
```

```python
#build a sample dataset containing similar distribution of normal trasction and fraud transaction

#number of fraud transaction =   177
```

```python
legit_sample = legit.sample(n=177)
```

```python
#concatenating two dataframes
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
# Convert relevant columns to numeric type before calculating mean
for col in new_dataset.columns:
    if new_dataset[col].dtype == 'object':
        try:
            new_dataset[col] = pd.to_numeric(new_dataset[col], errors='coerce')  # 'coerce' will replace non-numeric values with NaN
        except:
            pass  # Skip columns that can't be converted

new_dataset.groupby('Class').mean()
```

```python
#splitting data into features and targets
X= new_dataset.drop(columns='Class', axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#split the data inti training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
#random state =2
```

```python
print(X.shape, X_train.shape , X_test.shape)
```

```python
#Model training
#logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, Y_train)

```

```python
#Model Evaluation
#Accuracy Score
from sklearn.metrics import accuracy_score
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on traing data : ', training_data_accuracy)
#97% accuracy which is good !
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data: ', test_data_accuracy)
```
